### <span style="color:purple">Contingency Tables </span> 

```python
three_g_categories = {0: 'No 3G', 1: '3G'}
four_g_categories = {0: 'No 4G', 1: '4G'}
```

```python
df['three_g'] = df['three_g'].map(three_g_categories)
df['four_g'] = df['four_g'].map(four_g_categories)
```

```python
pd.crosstab(df['three_g'], df['four_g'],margins = True)
```

##### The above contingency table illustrates that 487 mobiles with 4G also have 3G.
##### There are 756 mobiles with 3G and 269 of them also have no 4G. 
##### Mobiles with no 3G or 4G are 244 mobiles

```python
wifi_x = {0: 'No Wifi', 1: 'Wifi'}

bluetooth_x = {0: 'No Bluetooth', 1: 'Bluetooth'}
```

```python
df['wifi'] = df['wifi'].map(wifi_x)

df['blue'] = df['blue'].map(bluetooth_x)
```

```python
pd.crosstab(df['blue'],df['wifi'])
```

##### Mobiles with No Wifi but with Bluetooth: 248 
##### Mobiles with No Wifi and No Bluetooth: 245
##### Mobiles with Wifi and Bluetooth (most repeated kind) :  268
##### Mobiles with Wifi but No Bluetooth : 239

```python
tab = pd.crosstab(df['three_g'],df['four_g'])
```

```python
tab1 = sm.stats.Table(tab)
print(tab1)
```

```python
tab1.fittedvalues.round(3)
# The coefficients represent the 'expected values'
# The number of phones "expected" to have 3G and 4G is 368.366
# The number of phones "expected" to have 4G but No 3G 119.134
# The number of phones "expected" to have No 4G but have 3G is 387.634
# The number of phones "expected" to have No 4G and No 3G is 125.366
```

```python
tab1.resid_pearson 
# The pearson residuals represent the standardized distances between the observed and expected events. 
# Since the coefficients are not close to 0, the variables 'four_g' and 'three_g' are dependents. 
```

```python
 n_test = tab1.test_nominal_association()
```

```python
print(n_test) 
```
